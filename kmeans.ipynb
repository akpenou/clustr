```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
from sklearn.cluster import KMeans, DBSCAN
```

```python
!ls -la images
```

```python
img = mpimg.imread('images/green_screen_test.png')
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
img_height, img_width, img_colors = img.shape
img_reshaped = np.reshape(img, (img_height * img_width, img_colors))
img_reshaped.shape
```

```python
model = KMeans(n_clusters = 2, random_state = 0)
model = model.fit(img_reshaped)
model.cluster_centers_
```

```python
new_img = np.reshape(model.labels_, (img_height, img_width))
new_img.shape
```

```python
plt.imshow(new_img)
```

```python
model = DBSCAN(eps=30, min_samples=50)
model = model.fit(img_reshaped)
model.core_sample_indices_
```

```python
new_img = np.reshape(model.labels_, (img_height, img_width))
new_img.shape
```

```python
plt.imshow(new_img)
```
